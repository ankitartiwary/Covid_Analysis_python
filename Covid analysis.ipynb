```python
covid_df.drop(["Sno" ,"Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace=True,axis =1)
```

```python
covid_df.head(5)
```

```python
covid_df["Date"]= pd.to_datetime(covid_df["Date"], format="%d-%m-%Y")
```

```python
covid_df.head()
```

```python
#Active cases

covid_df["Active_cases"] = covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
```

```python
covid_df.head()
```

```python
covid_df.tail()
```

```python
statewise =pd.pivot_table(covid_df, values = ["Confirmed","Deaths","Cured"], 
                          index= "State/UnionTerritory",
                         aggfunc=max)
```

```python
statewise.head()
```

```python
statewise["recovery rate"]=statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["mortality rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#top ten active cases states
top_10_active_cases = covid_df.groupby(by="State/UnionTerritory").max()[["Active_cases", "Date"]].sort_values(by= ["Active_cases"] ,ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))

plt.title("Top 10 active cases in India",size=26)

ax=sns.barplot(data=top_10_active_cases.iloc[:10],y='Active_cases',x=("State/UnionTerritory"),linewidth=3 ,edgecolor="black")
```

Maharashtra has most active cases

```python
#top states with highest death rate

top_10_deaths= covid_df.groupby(by="State/UnionTerritory").max()[["Deaths","Date"]].sort_values(by=["Deaths"],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,6))

plt.title("Top 10 death cases in India",size=26)

ax=sns.barplot(data=top_10_deaths.iloc[:12],y='Deaths',x="State/UnionTerritory",linewidth=2 ,edgecolor="black")
```

Due to more active case in Maharashtra ,death cases are equally high

```python
#remove rows where state =INDIA

vaccine = vaccine_df[vaccine_df.State !='India']
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated': 'Total'},inplace=True)
```

```python
vaccine.head()
```

```python
#most vaccinated state

max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending=False)[:10]

max_vac
```

```python
fig=plt.figure(figsize=(16,9))
plt.title('Top 10 Vaccinated states in India',size=25)
x=sns.barplot(data=max_vac.iloc[:10],y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='BLACK')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```

Maharashtra is most vaccinated state

```python
#least vaccinated state

min_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total',ascending=True)[:5]

min_vac
```

```python
fig=plt.figure(figsize=(16,5))
plt.title('Top 5 least Vaccinated states in India',size=25)
x=sns.barplot(data=min_vac.iloc[:10],y=min_vac.Total,x=min_vac.index,linewidth=2,edgecolor='BLACK')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```

Lakshdweep is least vaccinated state
